```python
import subprocess
import textwrap

rows = ws.get_all_values()
write_row = 2

MAX_CHARS = 45000  # safety buffer under Google limit

for i in range(1, len(rows)):
    raw_cell = rows[i][1]

    if not raw_cell:
        continue

    urls = raw_cell.split()

    for url in urls:
        if "youtube.com" not in url and "youtu.be" not in url:
            continue

        print("Processing:", url)

        result = subprocess.run(
            ["bun", "run", "src/cli.ts", "get", url],
            capture_output=True,
            text=True
        )

        transcript = result.stdout.strip()

        # 🔥 SPLIT INTO SAFE CHUNKS
        chunks = textwrap.wrap(transcript, width=MAX_CHARS)

        for idx, chunk in enumerate(chunks):
            ws.update_cell(write_row, 1, url)
            ws.update_cell(write_row, 2, chunk)
            ws.update_cell(write_row, 3, f"part_{idx+1}")

            write_row += 1
```

```python
import requests
import subprocess
import os
import csv
import io

# Your public sheet URL
url_of_sheet = 'https://docs.google.com/spreadsheets/d/1N5xFG0fkM7D_Ge1cZKjDzkg2jKeuzgWKhiPulVTMkyM/export?format=csv'

print("Downloading list from Google Sheets...")
response = requests.get(url_of_sheet)
response.raise_for_status()

# Read the CSV content
f = io.StringIO(response.text)
reader = csv.reader(f)
urls = [row[0] for row in reader if row and "youtube.com" in row[0]]

print(f"Found {len(urls)} URLs. Starting processing...")

# Ensure we are in the repo folder
if os.path.exists('ytranscript'):
    os.chdir('ytranscript')

# Batch processing
for url in urls:
    print(f"\n--- FETCHING: {url} ---")
    try:
        result = subprocess.run(
            ['bun', 'run', 'src/cli.ts', 'get', url, '--format', 'text'],
            capture_output=True,
            text=True,
            check=True
        )
        print(result.stdout)
    except Exception as e:
        print(f"Failed to process {url}: {e}")
```

```python
import subprocess

urls = [
"https://www.youtube.com/watch?v=a7VjpIqq8Xk",
"https://www.youtube.com/watch?v=DFSU-YPRTs8",
"https://www.youtube.com/watch?v=46t_cn0lx2E",
]

results = []

for url in urls:
    print(f"Processing: {url}")
    try:
        output = subprocess.check_output(["ytranscript", "get", url], text=True)
        results.append({"url": url, "transcript": output})
    except Exception as e:
        results.append({"url": url, "error": str(e)})

results[:2]
```
